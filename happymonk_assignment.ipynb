```python
df= pydataset.data("iris")
```

```python
df
```

```python
df.Species.value_counts()
```

```python
df1=df[(df.Species=='setosa')|(df.Species=='virginica')]
```

```python
df1.Species.value_counts()
```

```python
df1.shape
```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Activation, Dense
from sklearn.metrics import accuracy_score
```

```python
le = preprocessing.LabelEncoder()
labels = le.fit_transform(df1['Species'])
```

```python
df1['Species'] = labels
```

```python
df1.shape
```

```python
X = df1[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]]
y = df1[['Species']]

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
plt.scatter(df1['Petal.Width'], df1['Species'])

plt.show()
```

## KNN

```python
knnModel = KNeighborsClassifier(n_neighbors=2)
knnModel.fit(x_train, y_train)
```

```python
accuracy_score(y_test,knnModel.predict(x_test))
```

```python
print('the accuracy score for knn model is 100%')
```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Activation, Dense
from keras.layers import Dropout
from tensorflow.keras.utils import get_custom_objects
from tensorflow.python.keras.layers import LeakyReLU
from tensorflow.keras.layers import Activation
from tensorflow.keras.backend import sigmoid
```

## HIDDEN LAYER WITH SOFTMAX AND OUTLAYER WITH SIGMOID ACTIVATORS

```python
from tensorflow.keras.layers import BatchNormalization

# creating the ann as sequence of layers

ann1 = Sequential()

#adding first hidden layer with 30 neurons,the input layer will be added automatically

ann1.add(Dense(units=30,activation='softmax'))
ann1.add(BatchNormalization())
ann1.add(Dropout(0.5))

# adding output layer with 1 neuron, as this a binary classification

ann1.add(Dense(units=1,activation='sigmoid'))
```

```python
#compiling the ann using stochastic gradient descent (optimizer - adam)

ann1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

#setting callback for monitoring maximum accuracy

from tabnanny import verbose
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',mode = 'min',verbose=1,patience=25)

#traiining the ann with batch size of 32 (this is a batch learning)

model = ann1.fit(x_train,y_train,batch_size=32,validation_data=(x_test,y_test),epochs=500,callbacks=[early_stop])
```

```python
ss=ann1.predict(x_test)
```

```python
ann1pred=[]
for i in ss:
    if i>0.5:
     ann1pred.append(1)
    else:
        ann1pred.append(0)

ann1pred
```

```python
accuracy_score(y_test,ann1pred)
```

```python
print('the accuracy score for test data with hidden lay with softmax ann output layer with sigmoid is 100%')
```
